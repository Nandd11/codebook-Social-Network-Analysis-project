# NetworkX Social Network Analysis + Visuals



```python
from pathlib import Path
import networkx as nx
import matplotlib.pyplot as plt
from src.utils import read_json
from src.graph_builder import build_graph
from src.sna_metrics import compute_metrics, top_n
from src.community import detect_communities

users = read_json(Path('../data/cleaned_data.json'))
G = build_graph(users)
metrics = compute_metrics(G)
print('Nodes:', G.number_of_nodes(), 'Edges:', G.number_of_edges())
print('Top PageRank:', top_n(metrics['pagerank'], 5))

# Draw
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(8,6))
nx.draw(G, pos, node_size=80, with_labels=False)
plt.title('Social Graph')
plt.show()

comms = detect_communities(G)
print('Communities:', len(comms))
```
